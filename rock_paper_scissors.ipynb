```python
import random

def rps_game ():
    """
    user plays a single round of Rock, Paper, Scissors against the computer
    """
#start: define valid selections
options = ['rock', 'paper', 'scissors']

#user input and validation
while True:
    user_turn = input("Paper, rock, scissors. SHOOT! \n" \
    "Please input your selection.").lower()
    if user_turn in options:
        break #valid input exits the loop
    else:
        print("Invalid selection.\n" \
        "Please input 'rock', 'paper' or 'scissors' to complete your turn.")

#computer's choise
computer_turn = random.choice(options)

print(f"Your selection: {user_turn.capitalize()}")
print(f"Computer's selection: {computer_turn.capitalize()}")

#gamelogic and results
result = ""

if user_turn == computer_turn:
    result = "It's a tie!"

elif (user_turn == 'rock' and computer_turn == 'scissors') or \
(user_turn == 'scissors' and computer_turn == 'paper') or \
(user_turn == 'paper' and computer_turn == 'rock'):
    result = "You win!"

else:
    result = "Computer wins!"
#final output
    print(f"---Result: {result}---")
    
rps_game()
```
